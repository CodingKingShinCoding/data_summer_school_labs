```python
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
train_X = min_max_scaler.fit_transform(train_x_df)

Split_Ratio = 0.75
split_size = int(len(train_x_df)* Split_Ratio)
train_X = train_x_df[:split_size]
test_X = train_x_df[split_size:]

train_Y = y_train[:split_size]
test_Y = y_train[split_size:]
```

```python
test_X
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB,BernoulliNB
from sklearn.pipeline import Pipeline
from sklearn.cross_validation import KFold
from sklearn.cross_validation import cross_val_score
from sklearn import metrics
```

```python
clf_1 = LogisticRegression()
clf_2 = MultinomialNB()
clf_3 = BernoulliNB()
```

```python
pipe1 = Pipeline([['sc',MinMaxScaler()],['clf',clf_1]])
pipe2 = Pipeline([['sc',MinMaxScaler()],['clf',clf_2]])
pipe3 = Pipeline([['sc',MinMaxScaler()],['clf',clf_3]])


```

```python
def evaluate_cross_validatio(clf,X,Y,K):
    #k=5 교차검증
    cv = KFold(len(Y), n_folds=K, shuffle=True, random_state=0)
    
    scores = cross_val_score(clf,X,Y,cv=cv, scoring='roc_auc')
    print (scores)
    print('Mean score: %.3f +/- %.3f' % (np.mean(scores),np.std(scores)))
```

```python
clfs=[clf_1,clf_2,clf_3]

for clf in clfs:
    evaluate_cross_validatio(clf,train_X,train_Y,5)
```

```python
def train_and_evaluate(clf,X_train,X_test,Y_train,Y_test):
    clf.fit(X_train,Y_train)
    print ("Accuracy training set:")
    print(clf.score(X_train,Y_train))
    print ("Accuracy testing set:")
    print(clf.score(X_test,Y_test))
    Y_pred = clf.predict(X_test)
    
    
    print ("Classification Report")
    print (metrics.classification_report(Y_test,Y_pred))
    print ("Confusion Matrix")
    print (metrics.confusion_matrix(Y_test,Y_pred))
```

```python
train_and_evaluate(pipe1,train_X,test_X,train_Y,test_Y)
```
