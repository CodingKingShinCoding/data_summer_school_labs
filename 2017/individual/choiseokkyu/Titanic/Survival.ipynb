```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
df = pd.read_csv("train.csv")
df
```

### 필요없는 Columns 없애기

```python
df = df.drop("PassengerId", axis=1)
df = df.drop("Cabin", axis=1)
df = df.drop("Ticket", axis=1)
df = df.drop("Name", axis=1)
df = df.drop("Sex", axis=1)
df = df.drop("Fare", axis=1)
```

```python
df = pd.get_dummies(df)
df
```

### NaN값 채우기

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

### Logistic Regression

```python
y_data = df.pop("Survived")
```

```python
x_data = df
```

```python
from sklearn import linear_model, datasets

logreg = linear_model.LogisticRegression(fit_intercept=True)
logreg.fit(x_data, y_data)
```

```python
sum(logreg.predict(x_data) == y_data)  / len(y_data)
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_data, logreg.predict(x_data)))
```

### Naive Bayesian

```python
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.model_selection import cross_val_score
```

```python
MNB = MultinomialNB()
MNB.fit(x_data, y_data)

BNB = BernoulliNB()
BNB.fit(x_data, y_data)
```

```python
scores_M = cross_val_score(MNB, x_data, y_data, cv=5)
print("Multinomial Accuracy: %0.2f ± %0.2f" % (scores_M.mean(), scores_M.std() * 2))

scores_B = cross_val_score(MNB, x_data, y_data, cv=5)
print("Bernoulli Accuracy: %0.2f ± %0.2f" % (scores_B.mean(), scores_B.std() * 2))
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_data, MNB.predict(x_data)))
```

```python
print(classification_report(y_data, BNB.predict(x_data)))
```

### Roc Curve

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
scores = logreg.predict_proba(x_data)[:,1]
y_true = y_data
fpr_lr, tpr_lr, thresholds = roc_curve(y_true, scores)
roc_auc_lr = roc_auc_score(y_true, scores)
```

```python
import matplotlib.pyplot as plt

plt.figure()
lw = 2
plt.plot(fpr_lr, tpr_lr, color='darkorange',
        lw=lw, label='ROC curve (area = %0.2f)' % roc_auc_lr)

plt.plot([0,1],[0,1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```
