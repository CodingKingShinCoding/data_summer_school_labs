```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.io.parsers.read_csv('train.csv')
df_train.head()
```

```python
df_test = pd.io.parsers.read_csv('test.csv')
df_test.head()
```

```python
y_data = df_train.pop("SalePrice")
```

```python
df_train.shape, df_test.shape #데이터 크기 
```

```python
pd.get_dummies(df_train).head() #train의 dummies 뽑음 트레인이랑 테스트 컬럼수가다르다->concat으로
```

```python
pd.get_dummies(df_test).head()
```

```python
df_all = pd.concat([df_train, df_test])
```

```python
df_all.head() #나눠져있는 자료를 하나로 합친다. 
```

```python
df_all.isnull().sum() #데이터프레임 전부에서 null인 값들을 더해준다 null값이 너무 많은거 예를들어 Fence Alley같은건 날려버림
```

```python
for column_name in df_all.columns[df_all.isnull().sum() > 1000]: #직관적으로 쓰는걸 더 추천
    del df_all[column_name]
df_all.head()#null인 값들의 갯수를 더한게 1000보다 크면 지워라
#다섯개가 지워진것을 확인
```

```python
missing_values ={}
for column_name in df_all.columns:
    if df_all[column_name].dtypes == "object": #object인경우 모든 값을 찾아준다.
        #object type는 이름,속성(attribute), 함수명(method)에 해당
        mode_value = df_all[column_name].value_counts()[0] #최빈값
        missing_values[column_name] = mode_value#최빈값을 할당해줘라 
        df_all[column_name] = df_all[column_name].fillna(mode_value)#전부다 Nan값을 할당해줘라
    else:
        mean_value = df_all[column_name].mean() #평균값을 할당해줘라
        missing_values[column_name] = mean_value
        df_all[column_name] = df_all[column_name].fillna(mean_value)
```

```python
df_new_all = pd.get_dummies(df_all)#데이터하나하나가 무엇을 의미하는지 아는게 좋은편. 뭔값인지도 모르고 NOne이많은건 getdummies쓰기전에 날림
```

```python
list(df_new_all.columns)
#'BsmtQual_1298','BsmtQual_Ex','BsmtQual_Fa','BsmtQual_Gd'등생성확인
#참고로 빈칸은 알아서 없애줌, 의식적으로 없애주는것도 좋음 데이터를 뜯어서 찾아보는것도 중요
#사람들이 정확한 값을 넣는것이 아니니 하나하나 뜯어보고 날리고 해야함
```

```python
min_max_values = {a:[b,c]for a,b,c in zip(list(df_new_all.columns[1:]), df_new_all.max()[1:], df_new_all.min()[1:])}
min_max_values
#list의 column첫번째부터(column의 제목들만),df_new_all의 max값들, min값들 을 zip해준다(병렬정렬)-> list와 allmax allmin 값들이 하나의 tuple안에들어간다
#그러고나서 a,b,c,를 뽑아서 a:[b,c]에 맞게 넣어준다
#따로 최대,최소떼준이유는 스케일링할때 기존스케일링이랑 맞춰줘야함. 기준값들을 저장해줘야함. 
#
```

```python
for column_name in min_max_values.keys():
    df_new_all[column_name] = (df_new_all[column_name]-min_max_values[column_name][1])\
    /(min_max_values[column_name][0] - min_max_values[column_name][1]) * (5-1) + 1
#모든값들을 정규화 시켜준다
```

```python
x_train = df_new_all.iloc[:1460,1:].as_matrix()#saleprice전의값
x_test = df_new_all.iloc[1460:,1:].as_matrix() #Saleprice
y_train = y_data.as_matrix()#SalePrice
```

```python
from sklearn import linear_model
import matplotlib.pyplot as plt
regr = linear_model.LinearRegression()
regr.fit(x_train, y_train) #SalePrice전의 값들을x,Y값을 SalePrice로 잡고 둘사이 식을 구한다
#y ax+b의 식을 만들어준다.  ㄱ
#plt.plot(x_train,y_train,'x_train','y_train',1)
#plt.xlim(0,5)
#plt.ylim(0,12)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

mae = mean_absolute_error(y_train, regr.predict(x_train))#평균 절대오차
mse = mean_squared_error(y_train, regr.predict(x_train))#평균 제곱 오차
r2 = r2_score(y_train, regr.predict(x_train))

print(mae,mse,r2) #rsquare값은 1에 가까울수록 좋다. 밑에있는거에 제일 오른쪽에 있는 값!
```
